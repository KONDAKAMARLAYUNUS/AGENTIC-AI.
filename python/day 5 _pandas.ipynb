```python
def percentage():
 if('score'>90):
  return'90%'
 elif("score">60 and "score"<80):
  return'70%'
 else:
  return "fail"
  print(percentage(100))
```

```python
def grade(score):
  if score>=90:
    return "A"
  elif score>=80:
    return "B"
  elif score>=70:
    return "C"
  elif score>=60:
    return "D"
  else:
    return "F"
df['branch_letter']=df['score'].apply(grade)
print(df)
```

```python
pass_status={
    True:"pass",
    False:"fail"
}
df['result']=df['branch'].map(pass_status)
print(df)
```

```python
result={
    True:"pass",
    False:"fail"
}
df['result']=df['branch'].map(result)
print(df)
```
