```python
age=[1,2,3,4,5,6,7]
salary=[100,200,300,400,500,600,700]
id=[1000,2000,3000,4000,5000,6000,7000]
plt.figure(figsize=(5,5))
plt.plot(age,salary,marker="s",label="salary",color="#8836")
plt.plot(age,id, marker="^",label="id")
plt.legend();
plt.xlabel("age")
plt.ylabel("salary")
plt.grid(True,alpha=0.3)
plt.savefig("age_vry.png")
plt.show()
```

```python
cities=['mumbia','delhi','bangalore','chennai','hyderabad']
revenue=[4500,3000,5200,2900,3100]
fig,ax=plt.subplots(figsize=(10,5))
bar=ax.bar(cities,revenue,color='#0098')
```

```python
ages=[1,20,30,4,5,6,7,8,98,100]
plt.figure(figsize=(10,5))
plt.hist(ages,bins=100,color='#0098')
plt.scatter(ages,ages)
plt.axvline(np.mean(ages),color='red')
plt.axvline(np.median(ages),color='green')
plt.title("histogram")
plt.xlabel("age")
plt.ylabel("frequency")
plt.legend()
plt.show()
```
