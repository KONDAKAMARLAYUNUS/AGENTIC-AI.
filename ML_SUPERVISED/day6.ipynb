```python
data={
    'hours_studied':[1,2,3,4,5,6,7,8,9],
    'pass_fail':[0,0,3,2,4,5,6,7,8]

}
df=pd.DataFrame(data)


df['pass_fail'] = df['pass_fail'].apply(lambda x: 0 if x == 0 else 1)

x=df[['hours_studied']]
y=df['pass_fail']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(f'train: {len(x_train)}rows | Test: {len(x_test)}rows')

model=LogisticRegression()
model.fit(x_train,y_train)
print("model train sucessfully")
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f'accuracy: {accuracy}')


print(df)
```

```python
new_student=pd.DataFrame({'hours_studied': [2, 1, 13]})
prediction = model.predict(new_student)
print(f"Raw predictions: {prediction}")
labels={0:'fail',1:'pass'}
for hrs,pred in zip(new_student['hours_studied'],prediction):
  print(f' studied: {hrs} hrs->{labels[pred]}')
```
